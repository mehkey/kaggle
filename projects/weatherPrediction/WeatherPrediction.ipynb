```python
import requests
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Replace YOUR_API_KEY with your OpenWeatherMap API key
api_key = "c3bfbe561adc4bab893f3d5443af2751"

# Use the OpenWeatherMap API to retrieve forecast weather data for a specific location
location = "New York, US"
url = f"https://api.openweathermap.org/data/2.5/forecast?q={location}&appid={api_key}"

response = requests.get(url)
data = response.json()

# Extract the relevant information from the API response and store it in a list of dictionaries
weather_data = []
for item in data["list"]:
    date = item["dt_txt"]
    temperature = item["main"]["temp"]
    pressure = item["main"]["pressure"]
    humidity = item["main"]["humidity"]
    visibility = item["visibility"]
    cloud_cover = item["clouds"]["all"]
    wind_speed = item["wind"]["speed"]
    weather_data.append({"date": date, "temperature": temperature, "pressure": pressure, "humidity": humidity,
                         "visibility": visibility, "cloud_cover": cloud_cover, "wind_speed": wind_speed})

# Create a Pandas DataFrame from the list of dictionaries
df = pd.concat([pd.DataFrame(weather_data)], ignore_index=True)

# Use the RandomForestRegressor model to predict temperature based on humidity, pressure, visibility, cloud cover, and wind speed
X = df[["humidity", "pressure", "visibility", "cloud_cover", "wind_speed"]]
y = df["temperature"]

# Preprocess the input features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

# Evaluate the model's accuracy by comparing the predicted temperature to the actual temperature
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)


print(f"Accuracy: {accuracy * 100:.2f}%")
```
