```python
import pandas as pd
import psycopg2
```

```python
conn = psycopg2.connect(
    host='localhost',
    port=5432,
    database='postgres',
    user='postgres',
    password='password'
)


```

```python
cursor = conn.cursor()
```

```python
df = pd.read_csv('./instagram_profiles.csv', nrows=50,  delimiter=r"\s+",   on_bad_lines='skip')
#skiprows=1,
```

```python
df
```

```python
import sys
for index, row in df.iterrows():
    #print(row)
    if type(row['sid']) != int :
        continue

    #if row[0] > sys.maxsize:
    #    continue
    print(row)
    
    if row['sid'] < 1 or row['sid'] > 2_147_483_647:
        continue
        
    if not row['profile_name'] or not row['profile_id'] or not row['firstname_lastname']:
        continue
    
    cursor.execute(
      'INSERT INTO users (id, username, name, description) VALUES (%s, %s, %s, %s)',
      (row['sid'], row['profile_name'], row['profile_id'], row['firstname_lastname'])
    )


```

```python
cursor.execute('SELECT * FROM users')
rows = cursor.fetchall()

for row in rows:
    print(row[0], row[1], row[2], row[3])
```

```python
conn.commit()
cursor.close()
conn.close()
```
